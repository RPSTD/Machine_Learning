```python
# when you need to input simple letters as indexes instead of numbers, you need to give the indexes as "a","b","c". 
indexletter= {"uni":["pera","ruhunu","kdu","japura"],"faculty":["engi","law","management","tech"]}
indexletter
```

```python
# when you need to input simple letters as indexes instead of numbers, you need to give the indexes as "a","b","c".
ild=pd.DataFrame(indexletter,index=["a","b","c","d"])
ild
```

```python
# when using drop function, that will be remove the named index inside of the drop function. 
ilddrp=ild.drop("a")
ilddrp
```

```python
# when we drop multiples,
ilddrp2=ild.drop(["a","b"])
ilddrp2
```

```python
#when need to drop some value from orginal dataframe( according to this situation, ild), use inplace keyword as TRUE
ild.drop("a",inplace=True)
ild
```
